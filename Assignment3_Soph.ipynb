### What is the agentâ€™s score averaged over 1,000 games
##### The below suppress_stdout() function suppresses the print statements so that the 1000 games will run quickly

```python
import os
import sys
from contextlib import contextmanager

@contextmanager
def suppress_stdout():
    original_stdout = sys.stdout
    sys.stdout = open(os.devnull, 'w')
    try:
        yield
    finally:
        sys.stdout.close()
        sys.stdout = original_stdout

# Initialize a list to store the scores
scores = []

# Use the context manager to suppress print statements
with suppress_stdout():
    for _ in range(1000):
        env = Environment(4, 4, True, 0.2)
        prob_agent_game = ProbAgent(env)
        score = prob_agent_game.run()
        scores.append(score)

# Compute the average score
average_score = sum(scores) / len(scores)

# Print the average score
print(f"Average score over 1,000 games: {average_score}")
```
